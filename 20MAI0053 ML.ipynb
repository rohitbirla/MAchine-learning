```python
import numpy as np 
import pandas as pd 
```

```python
crx_data=pd.read_csv("crx.data.csv")
crx_data.head()
```

```python
crx_data.shape
```

**In the dataset crx.data, name the columns as col1, col2..col16**

```python
string="col"
col_list=[]
for i in range(1,17):
  new_string=string+str(i)
  col_list.append(new_string)

crx_data.columns=col_list
crx_data.head()
```


**Print the last 10 rows of data**

```python
crx_data.tail(10)
```

**Replace the ‘?’ with Not-a-Number**

```python
for cols in col_list:
  crx_data[crx_data[cols]=='?']=np.nan 
crx_data.head()
```

**Comment on the datatype of variables**

```python
crx_data.info()
```

**The col16 has + and -, replace them ‘P’ and ‘N’ respectively**

```python
crx_data['col16'].value_counts()
```

```python
crx_data=crx_data.replace({'col16': {'+': 'P', '-': 'N'}})
crx_data.head()	
```

```python
crx_data['col16'].value_counts()
```


**Find and display the number of variables of type ‘Object’**
